```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
xpoints = np.array([0, 5])
ypoints = np.array([0, 255])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,10])
ypoints = np.array([5, 15])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# Multiple points
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Markers
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# You can use also use the shortcut string notation parameter to specify the marker. 
# This parameter is also called fmt, and is written with this syntax:
# marker|line|color
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o--c')
plt.show()
```

```python
# Marker size
plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
# Marker Color
plt.plot(ypoints, marker = '*', ms = 15, mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = "o", ms = 15, mfc = 'r')
plt.show()
```

```python
# Linestyle
plt.plot(ypoints, linestyle = "dotted")
plt.show()
```

```python
# Line Color
plt.plot(ypoints, ls = "--", color= 'r')
plt.show()
```

```python
# Line Width
plt.plot(ypoints, linewidth = "11.11")
plt.show()
```

```python
# Multiple Lines
y1 = np.array([3, 6, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
# Create labels for a plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Set Font Properties for Title and Labels
font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports watch data", fontdict=font1, loc = "left") #loc - position the title
plt.xlabel("Average pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# Specify Which Grid Lines to Display
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
# Set Line Properties for the Grid
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid(color= "green", linestyle = '--', linewidth = 1)
plt.show()
```

```python
# With the subplot() function you can draw multiple plots in one figure:

# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.suptitle("My Shop")
plt.title("INCOME")
plt.show()
```

```python
# Creating Scatter Plots
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# Draw two plots on the same figure:

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x, y)
plt.show()
```

```python
# Set your own size for the markers
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s = sizes)
plt.show()
```

```python
# You can adjust the transparency of the dots with the alpha argument.

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s = sizes, alpha = 0.5)
plt.show()
```

```python
# Combine Color Size and Alpha
x = np.random.randint(100, size =(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size = (100))
sizes = 10*np.random.randint(100, size=(100))

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5, cmap = 'nipy_spectral')
plt.colorbar()
plt.show()

```

```python
# Bars
x  = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y)
plt.show()
```

```python
# Horizontal bar

x  = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
# Bar color

x  = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = 'r')
plt.show()
```

```python
# Bar width
x  = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
# Bar height

x  = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
# Explode

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
# Legend
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four fruits")
plt.show() 
```
